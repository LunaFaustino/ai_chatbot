```python
import matplotlib.pyplot as plt
```

```python
# plot para gráfico de linha

plt.plot(
    [2,9,9,2,2],
    [2,2,9,9,2]
)
```

```python
#scatter para gráfico de ponto

plt.scatter(
    [2,9,9,2,2],
    [2,2,9,9,2],
    label="Pontos do cubo", #legenda dos pontos
    c="blue", #cor dos pontos
    edgecolors="red" #bordas dos pontos
)

plt.title("Título")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend(loc=3, bbox_to_anchor=(1,0.5)) #posiciona a legenda
```

```python
# dois gráficos em 1, sempre começa com Figure e termina com SHOW

plt.figure()
plt.plot(
    [2,9,9,2,2],
    [2,2,9,9,2],
    c="pink"
)

plt.scatter(
    [2,9,9,2,2],
    [2,2,9,9,2],
    c="blue"
)

plt.title("Gráfico sobre gráfico")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.tight_layout()
plt.savefig("quadrado.png")
plt.show()
```
